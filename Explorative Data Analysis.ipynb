```python
#step 1: import library
import pandas as pd
```

```python
#step 2: read data set
df = pd.read_csv("https://github.com/YBIFoundation/Dataset/raw/main/Salary%20Data.csv")
```

```python
#step 3: define output(salary) = (Y) and input(experience) = (X)
df.columns
```

```python
Y = df['Salary']
X = df[['Experience Years']]
```

```python
import seaborn as sns
sns.regplot(x='Experience Years', y='Salary', data=df, fit_reg=True)
```

```python
#step 4: select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#step 5: fit model
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#step 6: predict
y_pred = model.predict(X)
```

```python
#step 7: evaluate
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y,y_pred)
```

```python
print(mean_absolute_percentage_error(Y,y_pred)*100)
```

```python
accuracy = (100-mean_absolute_percentage_error(Y,y_pred))
print(accuracy)
```
